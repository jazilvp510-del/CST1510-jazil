### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
# Step 1: Ask the user for temperatures and write to file
with open("temps.txt", "w") as file:
    for day in range(1, 6):
        temp = input(f"Enter temperature for Day {day}: ")
        file.write(f"Day {day}: {temp}°C\n")

# Step 2: Read the file and calculate average
total_temp = 0
count = 0

with open("temps.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        # Each line format: "Day 1: 22°C"
        temp_str = line.split(":")[1].strip().replace("°C", "")
        temp_num = float(temp_str)
        total_temp += temp_num
        count += 1

average_temp = total_temp / count

# Step 3: Print average temperature
print(f"Average temperature over {count} days: {average_temp:.2f}°C")
`````

### **Activity 4: Movie Collection Tuple** 


Create a tuple with 3 movies and their release years:
movies = (("Inception", 2010), ("Matrix", 1999), ("Interstellar", 2014))

Print:
- The oldest movie
- All movie titles
- Try to change a year (show it causes an error)

---

`````python
# Step 1: Create the tuple
movies = (("Inception", 2010), ("Matrix", 1999), ("Interstellar", 2014))

# Step 2: Find the oldest movie
oldest = min(movies, key=lambda x: x[1])  # x[1] is the year
print("The oldest movie is:", oldest[0], "released in", oldest[1])

# Step 3: Print all movie titles
titles = [movie[0] for movie in movies]
print("All movie titles:", titles)

# Step 4: Try to change a year (will cause an error)
try:
    movies[0] = ("Inception", 2011)
except TypeError as e:
    print("Error:", e)
`````

### **Activity 5: Unique Hobbies Set** 


Two friends share their hobbies:
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

Find and print:
- Hobbies both enjoy (intersection)
- Hobbies only Alex enjoys
- All unique hobbies combined

---

`````python
# Step 1: Define the sets
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

# Step 2: Hobbies both enjoy (intersection)
common_hobbies = alex_hobbies & sam_hobbies  # or alex_hobbies.intersection(sam_hobbies)
print("Hobbies both enjoy:", common_hobbies)

# Step 3: Hobbies only Alex enjoys
alex_only = alex_hobbies - sam_hobbies  # or alex_hobbies.difference(sam_hobbies)
print("Hobbies only Alex enjoys:", alex_only)

# Step 4: All unique hobbies combined (union)
all_hobbies = alex_hobbies | sam_hobbies  # or alex_hobbies.union(sam_hobbies)
print("All unique hobbies combined:", all_hobbies)
`````


### **Activity 6: Simple Grade Book Dictionary** 

Create a dictionary for 3 students and their scores:
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

Perform these tasks:
- Add a new student "Diana" with score 88
- Update Bob's score to 95
- Print all students with scores above 80
- Calculate and print the average score

---

`````python
# Step 1: Create the dictionary
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

# Step 2: Add a new student
grades["Diana"] = 88

# Step 3: Update Bob's score
grades["Bob"] = 95

# Step 4: Print all students with scores above 80
print("Students with scores above 80:")
for student, score in grades.items():
    if score > 80:
        print(f"{student}: {score}")

# Step 5: Calculate and print the average score
average_score = sum(grades.values()) / len(grades)
print(f"Average score: {average_score:.2f}")
`````

### **Activity 7: Contact Book** 


Create a contact dictionary:
contacts = {
    "Mom": "555-1234",
    "Dad": "555-5678",
    "Friend": "555-9012"
}

Write a program that:
- Adds a new contact
- Looks up a contact by name
- Prints all contacts in format "Name: Number"

---

`````python
# Step 1: Create the initial contact dictionary
contacts = {
    "Mom": "555-1234",
    "Dad": "555-5678",
    "Friend": "555-9012"
}

# Step 2: Add a new contact
new_name = input("Enter the name of the new contact: ")
new_number = input("Enter the phone number: ")
contacts[new_name] = new_number

# Step 3: Look up a contact by name
lookup_name = input("Enter a name to look up: ")
if lookup_name in contacts:
    print(f"{lookup_name}: {contacts[lookup_name]}")
else:
    print(f"{lookup_name} not found in contacts.")

# Step 4: Print all contacts
print("\nAll contacts:")
for name, number in contacts.items():
    print(f"{name}: {number}")
`````
